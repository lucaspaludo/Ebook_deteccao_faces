# Exemplo 1

```python
import cv2
import matplotlib.pyplot as plt
import dlib

imagem = cv2.imread('../Images/people4.png')
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(imagem_rgb)
plt.show()

```

```python
detector_face_cnn = dlib.cnn_face_detection_model_v1('Weights/mmod_human_face_detector.dat')

```

```python
deteccoes = detector_face_cnn(imagem_rgb)

for face in deteccoes:
    l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
    print(c)
    cv2.rectangle(imagem_rgb, (l, t), (r, b), (0, 255, 255), 2)

plt.axis('off')
plt.imshow(imagem_rgb)
plt.show()
```
