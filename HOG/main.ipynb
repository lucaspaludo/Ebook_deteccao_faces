# Detecção de faces com HOG

# Carregamento da imagem e pré-processamento

```python
# importanto bibliotecas que iremos utilizar
import cv2
import matplotlib.pyplot as plt
import dlib

# realizando leitura da imagem
imagem = cv2.imread('../Images/people4.png')

#exibindo imagem sem processamento
plt.axis('off')
plt.imshow(imagem)
plt.show()
```

```python
# convertendo imagem para o formato RGB
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(imagem_rgb)
plt.show()
```

```python
#exibindo resolução da imagem
imagem_rgb.shape
```

```python
# Retorna um detector de faces baseado em HOG (Histogram of Oriented Gradients) 
# treinado previamente pela biblioteca dlib
detector_face_hog = dlib.get_frontal_face_detector()
```

```python
#Ela executa o detector HOG sobre a imagem e retorna as faces encontradas
deteccoes = detector_face_hog(imagem_rgb)
```

```python
# Retorna as coordenadas das faces encontradas e quantas 
# foram encontradas no total
deteccoes, len(deteccoes)
```

```python
# exibindo imagem com as detecções na imagem colorida
for face in deteccoes:
    l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
    cv2.rectangle(imagem_rgb, (l, t), (r, b), (255, 255, 0), 4)

plt.axis('off')
plt.imshow(imagem_rgb)
plt.show()
```
